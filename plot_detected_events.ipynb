```python
E = np.linspace(1, 100, 1000000) # Neutrino detection energy
detector = 'Super-k'
resolution = 0.05 # Fraction of error in energy reconstruction
distance = 10
```

**Resolution of DUNE** $\approx 0.2$, independent of neutrino energy

Abi, B., Acciarri, R., Acero, M. A., Adamov, G., Adams, D., Adinolfi, M., ... & Childress, S. (2020). Deep underground neutrino experiment (dune), far detector technical design report, volume ii: Dune physics. arXiv preprint arXiv:2002.03005.

```python
ref = energy_sampler(E, 5e52, resolution, 'proportional', detector, 'normal', distance, phi = 0)
ref_invert = energy_sampler(E, 5e52, resolution, 'proportional', detector, 'inverted', distance, phi = 0)
```

```python
y_e = detection_spectra(E, 5e52, flavor = 'nu_e', detector = detector, hierarchy = 'normal', distance = distance, phi = 0)
y_ebar = detection_spectra(E, 5e52, flavor = 'nubar_e', detector = detector, hierarchy = 'normal', distance = distance, phi = 0)
y_x = detection_spectra(E, 5e52, flavor = 'nu_x', detector = detector, hierarchy = 'normal', distance = distance, phi = 0)

y_e2 = detection_spectra(E, 5e52, flavor = 'nu_e', detector = detector, hierarchy = 'inverted', distance = distance, phi = 0)
y_ebar2 = detection_spectra(E, 5e52, flavor = 'nubar_e', detector = detector, hierarchy = 'inverted', distance = distance, phi = 0)
y_x2 = detection_spectra(E, 5e52, flavor = 'nu_x', detector = detector, hierarchy = 'inverted', distance = distance, phi = 0)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,6))
ax[0].hist(ref['Total'], bins = [i for i in range(100)], histtype = 'step', color = 'black',
        lw = 2)
ax[0].plot(E, y_e+y_ebar+y_x, ls = '--', color = 'purple', lw = 1.5)
ax[0].plot([80, 100], [0,10], label = r'Eventos totais', color = 'black', lw = 2)
ax[0].hist(ref['nu_e'], bins = [i for i in range(100)], histtype = 'step', color = 'crimson',
        lw = 2)
ax[0].plot([80, 100], [0,10], label = r'$\nu_e$', color = 'crimson', lw = 2)
ax[0].hist(ref['nubar_e'], bins = [i for i in range(100)], histtype = 'step', color = 'teal',
        lw = 2)
ax[0].plot([80, 100], [0,10], label = r'$\overline{\nu}_e$', color = 'teal', lw = 2)
ax[0].hist(ref['nu_x'], bins = [i for i in range(100)], histtype = 'step', color = 'orange',
        lw = 2)
ax[0].plot([80, 100], [0,10], label = r'$\nu_x$', color = 'orange', lw = 2)
lgd = ax[0].legend(title = 'Tipo de neutrino', loc = 'upper right', fontsize = 14)
title = lgd.get_title()
title.set_fontsize(14)
ax[0].set_xlabel('Energia do neutrino [MeV]', fontsize = 16)
ax[0].set_ylabel('Eventos', fontsize = 16)
ax[0].set_title(f'Ordenamento normal', fontsize = 18)
ax[0].set_xlim(0, 70)
ax[0].set_ylim(0, 500)
# plt.yticks(fontsize = 14)
# plt.xticks(fontsize = 14)

ax[1].hist(ref_invert['Total'], bins = [i for i in range(100)], histtype = 'step', color = 'black',
        lw = 2)
ax[1].plot(E, y_e2+y_ebar2+y_x2, ls = '--', color = 'purple', lw = 1.5)
ax[1].plot([80, 100], [0,10], label = r'Eventos totais', color = 'black', lw = 2)
ax[1].hist(ref_invert['nu_e'], bins = [i for i in range(100)], histtype = 'step', color = 'crimson',
        lw = 2)
ax[1].plot([80, 100], [0,10], label = r'$\nu_e$', color = 'crimson', lw = 2)
ax[1].hist(ref_invert['nubar_e'], bins = [i for i in range(100)], histtype = 'step', color = 'teal',
        lw = 2)
ax[1].plot([80, 100], [0,10], label = r'$\overline{\nu}_e$', color = 'teal', lw = 2)
ax[1].hist(ref_invert['nu_x'], bins = [i for i in range(100)], histtype = 'step', color = 'orange',
        lw = 2)
ax[1].plot([80, 100], [0,10], label = r'$\nu_x$', color = 'orange', lw = 2)
lgd = ax[1].legend(title = 'Tipo de neutrino', loc = 'upper right', fontsize = 14)
title = lgd.get_title()
title.set_fontsize(14)
ax[1].set_xlabel('Energia do neutrino [MeV]', fontsize = 16)
ax[1].set_ylabel('Eventos', fontsize = 16)
ax[1].set_title(f'Ordenamento inverso', fontsize = 18)
ax[1].set_xlim(0, 70)
ax[1].set_ylim(0, 500)

fig.suptitle(f'Espectro de detecção {detector} para supernova a {distance} Kpc', fontsize = 22)
fig.patch.set_alpha(0.4)
# plt.savefig(f'Plots/detection_spectrum_{detector}_{distance}_kpc_pt-br.svg',
#             bbox_inches = 'tight', dpi = 300)
plt.show()
```

```python
detectors = ['Super-k', 'DUNE', 'JUNO', 'Hyper-k']
resolutions = [0.05, 0.1, 0.05, 0.03] # Fraction of error in energy reconstruction

for k in range(4):
    detector = detectors[k]
    resolution = resolutions[k]
    
    results = {}
    distances = []
    means_no = []
    means_io = []
    std_no = []
    std_io = []
    for i in tqdm(range(50)):
        if i == 0:
            distance = 0.1
        else:
            distance = 1 + 2*i
        tot_no = []
        tot_io = []
        for j in range(30):
            ref = energy_sampler(E, 5e52, resolution, 'proportional', detector, 'normal', distance, phi = 0, print_expected=False, only_total_events=True)
            ref_invert = energy_sampler(E, 5e52, resolution, 'proportional', detector, 'inverted', distance, phi = 0, print_expected=False, only_total_events=True)
            tot_no.append(ref['Total'])
            tot_io.append(ref_invert['Total'])
        distances.append(distance)
        means_no.append(np.mean(tot_no))
        means_io.append(np.mean(tot_io))
        std_no.append(np.std(tot_no))
        std_io.append(np.std(tot_io))
        
    distances = np.array(distances)
    means_no = np.array(means_no)
    means_io = np.array(means_io)
    std_no = np.array(std_no)
    std_io = np.array(std_io)
    
    results['Distances'] = distances
    results['Mean N normal'] = means_no
    results['Mean N inverted'] = means_io
    results['SD N normal'] = std_no
    results['SD N inverted'] = std_io
    
    pd.DataFrame(results).to_csv(f'Total_events_detected_{detector}.csv')
```

```python
fig, ax = plt.subplots(figsize=(10,8))
detectors = ['Super-k', 'DUNE', 'JUNO', 'Hyper-k']
colors = ['#fff433','#ffffff','#9b59d0','#000000']
ymin = 50
ymax = 1e9

for i in range(4):
    df = pd.read_csv(f'Total_events_detected_{detectors[i]}.csv')
    plt.plot(df['Distances'], df['Mean N normal'], label = f'{detectors[i]} (NO)',
             lw = 2, color = colors[i])
    plt.plot(df['Distances'], df['Mean N inverted'], label = f'{detectors[i]} (IO)',
             lw = 2, ls = '--', color = colors[i])
    plt.fill_between(df['Distances'], df['Mean N normal']-2*df['SD N normal'],
                     df['Mean N normal']+2*df['SD N normal'], alpha = 0.2,
                    color = colors[i])
    plt.fill_between(df['Distances'], df['Mean N inverted']-2*df['SD N inverted'],
                     df['Mean N inverted']+2*df['SD N inverted'], alpha = 0.2,
                    color = colors[i])

plt.yscale('log')
plt.yticks(fontsize = 16)
plt.xticks(fontsize = 16)
plt.ylim(ymin, ymax)
plt.xscale('log')
plt.xlim(0.1, 100)
lgd = plt.legend(bbox_to_anchor=(1.27,1.015), ncol = 1, title = 'Detectors', fontsize = 14)
plt.setp(lgd.get_title(), fontsize = 16)
frame = lgd.get_frame()
frame.set_facecolor('lightgrey')
plt.xlabel('Distance from Earth [kpc]', fontsize = 22)
plt.ylabel(r'Total detections $N_\nu$', fontsize = 22)
plt.title(r'Number of detected neutrinos from a CCSN', fontsize = 26, pad = 28)
fig.text(0.5, 0.89, r'Total energy $L_T = 1.5 \times 10^{53}$ ergs', fontsize = 20, ha = 'center')
fig.text(0.6, -0.06, r'$N_\nu = \displaystyle \sum_{f \in \{\nu_e, \overline{\nu}_e, \nu_x\}} \frac{L_T N_f^{\mathrm{target}} (1+\alpha_f)^{1+\alpha_f}}{4 \pi D^2 \Gamma(1+\alpha_f) \left \langle E_f \right \rangle^{2+\alpha_f}} \displaystyle \int_0^\infty \mathcal{F}_f^d(E) \sigma_f(E) \varepsilon(E) dE$', fontsize = 22,
        ha = 'center')
plt.vlines(8, ymin, ymax, ls = '--', color = 'black', lw = 1)
plt.vlines(16, ymin, ymax, ls = '--', color = 'black', lw = 1)
plt.vlines(0.197, ymin, ymax, ls = '--', color = 'black', lw = 1)
plt.text(7.8, 1e7, 'Galactic center', fontsize = 16, ha = 'right')
plt.text(0.21, 2e5, 'Betelgeuse', fontsize = 16)
plt.text(16.5, 2e5, 'Galaxy edge', fontsize = 16)
fig.add_artist(mpl.lines.Line2D([0, 1.1], [0.03, 0.03], color = 'black', lw = 1, ls = '--'))

fig.patch.set_alpha(1)
# fig.patch.set_color('grey')
for ax in fig.axes:
    ax.patch.set_color('lightgrey')
plt.savefig('Plots/Detected_events.svg', dpi = 300, bbox_inches = 'tight')
plt.savefig('Plots/Detected_events.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
