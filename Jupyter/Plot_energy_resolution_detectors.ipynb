```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
E = np.linspace(0, 100, 1000)

dune = 0.11*np.sqrt(E) + 0.02*E
super_k = 0.5*np.sqrt(E)
juno = 0.03*np.sqrt(E)
```

```python
fig, ax = plt.subplots(figsize=(8,4))

plt.plot(E, super_k, lw = 2, color = sns.color_palette('inferno')[0], label = 'Super/Hyper-k')
plt.plot(E, dune, lw = 2, color = sns.color_palette('inferno')[2], label = 'DUNE')
plt.plot(E, juno, lw = 2, color = sns.color_palette('inferno')[4], label = 'JUNO')
plt.xlim(0, 100)
plt.xticks(fontsize = 12)
plt.xlabel(r'Neutrino energy $E_\nu$ [MeV]', fontsize = 14)
plt.ylim(0, 5)
plt.yticks(fontsize = 12)
plt.ylabel(r'Energy resolution $\delta$ [MeV]', fontsize = 14)
plt.legend(loc = 'upper left', fontsize = 12)
plt.title('Detectors energy resolution', fontsize = 18, pad = 10)

plt.savefig('Plots/Energy_resolution_detectors.png', dpi = 300, bbox_inches = 'tight')
plt.savefig('Plots/Energy_resolution_detectors.svg', dpi = 300, bbox_inches = 'tight')
plt.show()
```
